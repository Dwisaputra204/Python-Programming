```python
def cek_plat_nomor(tanggal, plat_nomor):
    plat_nomor = plat_nomor.split(' ')
    plat_nomor = int(plat_nomor[1])
    if tanggal % 2 == 0:
        if plat_nomor % 2 ==0:
            print('Anda lolos')
        else:
            print('Anda ditilang! Hari ini tanggal genap, plat nomor Anda ganjil')
    else:
        if plat_nomor % 2 == 0:
            print('Anda ditilang! Hari ini tanggal ganjil, plat nomor Anda genap')
        else:
            print('Anda lolos')
```

## Default Argument

```python
def total_gaji(gaji_pokok, bonus = 0, pajak = 0.1, denda = 0):
    total_gaji = gaji_pokok + bonus - denda - (gaji_pokok * pajak)
    result = f'total_gaji : {int(total_gaji)}' # format string
    return result
```

```python
total_gaji(6000000)
```

```python
gaji_januari = 6000000
bonus_januari = 500000

total_gaji(gaji_januari, bonus_januari)
```

## Variable-Length Argument

```python
def total_bilangan(*bilangan, **keywords):
    hitung = 0
    for bil in bilangan:
        hitung += bil
    for key in keywords:
        hitung += keywords[key]
    print(bilangan)
    print(keywords)
    return print('hasilnya adalah : ',hitung)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    print(x)
    newlist.append(x)

print(newlist)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

newlist = [x for x in fruits if "a" in x]

print(newlist)
```

```python
# absolute value
def absolute_value(num):
    'fungsi ini akan mengembalikan nilai absolute'
    
    if type(num) is not int and type(num) is not float:
        return print('error: masukkan angka')
    elif num >= 0:
        return num
    else:
        return -num
```

```python
absolute_value(2)
```

```python
absolute_value(-2.0)
```

```python
absolute_value('asdf')
```

```python
absolute_value([1,2,3])
```

```python
# BMI
# BMI = berat (Kg) / tinggi^2 (m^2)
def bmi(berat, tinggi):
    '''
    Menghitung indeks massa tubuh/ Body Mass Index.
    Berat dalam Kg
    Tinggi dalam cm (diubah secara otomatis kedalam meter)
    '''
    tinggi /= 100
    bmi = berat/tinggi**2
    return bmi
```

```python
bmi(60, 170)
```

```python
# Iterasi Total Gaji Tiap Bulan
def total_gaji(gaji_pokok, bonus = 0, pajak = 0.1, denda = 0):
    total_gaji = gaji_pokok + bonus - denda - (gaji_pokok * pajak)
    result = int(total_gaji)
    return result
```

```python
bonus_dict = {'Januari' : 600000, 'Februari' : 500000, 'Maret' : 400000, 'April' : 800000}
gaji_pokok = 6000000

for bulan, bonus in bonus_dict.items():
    total = total_gaji(gaji_pokok, bonus)
    print('Total gaji bulan', bulan, 'adalah', total)
```
